```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_excel('data.xlsx', engine='openpyxl')
df.head()
```

```python
df = df.drop(['Unnamed: 5'], axis=1) #Удаление пустого столбца
df = df.drop(labels = [0,1],axis = 0) # Удаление '-', не уверен, что это правильное решение, но непонятно к какому промежуткому времени относятся, поэтому решил удалить данную строчку 
df['receiving_date'] = pd.to_datetime(df['receiving_date']) # STR -> DATE format
```

```python
jun = df.query("'2021-06-01' < receiving_date < '2021-06-30'")
jul = df.query("'2021-07-01' < receiving_date < '2021-07-31'")
aug = df.query("'2021-08-01' < receiving_date < '2021-08-31'")
sep = df.query("'2021-09-01' < receiving_date < '2021-09-30'")
oct = df.query("'2021-10-01' < receiving_date < '2021-10-31'")
nov = df.query("'2021-11-01' < receiving_date < '2021-11-30'")
```

```python
juns = jun['sum'].sum()
juls = jul['sum'].sum()
augs = aug['sum'].sum()
seps = sep['sum'].sum()
octs = oct['sum'].sum()
novs = nov['sum'].sum()
res_2 = pd.DataFrame({'month': ['jun', 'jul', 'aug','sep','oct','nov'], 'sum': [juns, juls, augs,seps,octs,novs]})
res_2
```

```python
plot = res_2.sort_values(by='month', axis=0, ascending=False)
fig = px.bar(res_2, 
             x='month', 
             y='sum',
            )

fig.show()
```
